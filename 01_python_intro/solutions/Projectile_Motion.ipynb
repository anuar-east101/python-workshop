# Projectile Motion

First, let's import the libraries that we need.

```python
import matplotlib.pyplot as plt
import numpy as np
import math

import imageio
from IPython.display import Image
import glob, os
```

Create a function called `trajectory` that receives the following variables:
1. `t`: time
2. `alpha`: initial angle of projection
3. `V`: initaial velocity of the object
4. `g`: gravity

In this `trajectory` function, perform the projectile motion calculation using the following equations:

$x = Vt cos (\alpha)$

$y = Vt sin (\alpha) - \frac{1}{2}gt^2$

Then return the calculated values of `x` and `y`.

```python
def trajectory(t, alpha, V, g):
    
    x = V * t * math.cos(alpha)
    y = V * t * math.sin(alpha) - 1/2 * g * t**2
    
    return x, y
```

Define `V`, `alpha`, and `g` to be 100 $m/s$, $\pi/2$ (in radian), and 9.8 $m^2/s$ respectively.

```python
V = 100 
alpha = math.pi/4
g = 9.8
```

We are going to run the calculation from $t=0$ sec to $t=15$ sec. 

Create an array called `times` that has 30 points using `np.linspace`.

```python
times = np.linspace(0,15,30)
```

Run the code below to generate an animation of the projectile motion from time $t=0$ to $t=15$ sec.

```python
images = [] # initialise a placeholder for the images to be converted to gif/video

# iterate for all t in times array
for t in times:
    
    x, y = trajectory(t, alpha, V, g)
    
    plt.figure(figsize=(8,5))
    
    plt.scatter(x, y, c='g', s=100)
    plt.xlabel('Distance, $x$ (m)')
    plt.ylabel('Height, $y$ (m)')
    plt.title('t = ' + str(round(t,2)) + ' sec')
    
    plt.xlim(0,1200)
    plt.ylim(0,300)
    
    file_name = 'plot_'+str(round(t))+'.png'
    
    plt.savefig(file_name)
    plt.close()
    
    # append the new figures into the placeholder
    images.append(imageio.imread(file_name))
    
imageio.mimsave('trajectory.gif', images)

# delete the redundant images used to create the gif
for f in glob.glob("plot_*.png"):
    os.remove(f)  
```
